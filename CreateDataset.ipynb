Ignorando um tipo de diamante para o modelo de regressão logística permanecer binário (primeira versão)

```python
df = df[df['Type'].isin(['GIA', 'IGI Lab-Grown'])]
```

```python
df['Type'].value_counts()
```

```python
df.info()
```

```python
df['Type'].value_counts()
```

```python
df['Train'] = 0
```

```python
df.loc[df[df['Type'] == 'IGI Lab-Grown'].index[:500], 'Train'] = 1
df.loc[df[df['Type'] == 'GIA'].index[:500], 'Train'] = 1
```

```python
df['Train'].value_counts()
```

```python
df[df['Train'] == 1]
```

```python
df_train = df[df['Train'] == 1].drop(['Train'], axis=1)
```

```python
df_train.to_csv('data/diamonds_train.csv', index=False)
```

```python
df_validate = df[df['Train'] == 0].drop(['Train'], axis=1)
```

```python
df_validate.to_csv('data/diamonds_validate.csv', index=False)
```
